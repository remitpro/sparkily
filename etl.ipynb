```python
import os
import glob
import pandas as pd
```

```python
song_dir = 'data/song_data'
log_dir = 'data/log_data'
```

```python
song_files = []
```

```python
for root, dirs, files in os.walk(song_dir):
        files = glob.glob(os.path.join(root,'*.json'))
        for f in files :
            song_files.append(os.path.abspath(f))


num_files = len(song_files)
print('{} files found in {}'.format(num_files, song_dir))
```

```python
for i, datafile in enumerate(song_files):
    print('Processing file {}/{}: {}'.format(i + 1, num_files, datafile))
    df = pd.read_json(datafile, lines=True)
    if i == 0:
        song_df = df
    else:
        song_df = pd.concat([song_df, df], ignore_index=True)
```

```python
song_df.head
```

```python
song_df.to_csv('sparkily/seeds/song_data.csv', index=False)
print('Song data saved to sparkily/seeds/song_data.csv')
```

```python
log_files = []
for root, dirs, files in os.walk(log_dir):
    files = glob.glob(os.path.join(root, '*.json'))
    for f in files:
        log_files.append(os.path.abspath(f))

num_files = len(log_files)
print('{} files found in {}'.format(num_files, log_dir))
```

```python
for i, logfile in enumerate(log_files):
    print('Processing file {}/{}: {}'.format(i + 1, num_files, logfile))
    df = pd.read_json(logfile, lines=True)
    if i == 0:
        log_df = df
    else:
        log_df = pd.concat([log_df, df], ignore_index=True)
```

```python
log_df.head()
```

```python
log_df.shape
```

```python
# convert registration column to string
log_df['registration'] = pd.to_datetime(log_df['registration'], unit='ms')

# convert ts column to datetime
log_df['ts'] = pd.to_datetime(log_df['ts'], unit='ms')

log_df.head()
```

```python
#write to csv 
log_df.to_csv('sparkily/seeds/log_data.csv', index=False)
print('Log data saved to sparkily/seeds/log_data.csv')
```
